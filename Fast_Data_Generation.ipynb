```python
# Dependencies
import numpy as np
import pandas as pd
from faker import *
fake = Faker()
```

```python
NUM_ROWS =  50000
customer_id_range = [i for i in range(100000,999999)]
age_range = [i for i in range(18,65)]

## You can set your values in a variable and then call them in numpy random choice function and set the size as to how many rows you want, 
customer_id = np.random.choice(customer_id_range, replace = False,size = NUM_ROWS)
age = np.random.choice(age_range,size = NUM_ROWS)

## You can put random values straight up without any variable by using numpy random randint function
income_estimate = np.random.randint(90000,120000,size = NUM_ROWS)

## You need random numbers from 0 uptill 1, use random.random function of numpy
dti_ratio = np.random.random(size = NUM_ROWS)

## You need random numbers with set probability distribution of these numbers, use p in random choice
risk_seg = ['C', 'F', 'H', 'M', 'L', 'S'] # C(Super High), F(Elevated), H(High), M(Med), L(Low) ,or S(Super Low)
risk = np.random.choice(risk_seg, p = [0.05,0.1,0.35,0.35, 0.1, 0.05],size = NUM_ROWS) # C will be 0.05 percent of total records generated

gender =  np.random.choice([0,1], p = [0.4,0.6],size = NUM_ROWS)

df_cust = pd.DataFrame({'customer_id' : customer_id, 'age':age,'gender':gender,'risk':risk,'income_estimate':income_estimate,'dti_ratio':dti_ratio})
df_cust
```

```python
# Write your own custom function
def email_generator(fname,lname):
  return f"{fname.lower()}.{lname.lower()}@{fake.domain_name()}"

df_cust['customer_id'] = 'CUST' + df_cust['customer_id'].astype(str)
# Converting Gender to Male and Female using numpy where function
df_cust['gender'] = np.where(df_cust.gender.values == 0, 'M', 'F')
# Using faker to generate random first and last name using pandas apply function, works just as fast
df_cust['fname'] = df_cust.apply(lambda x: fake.first_name_male() if x['gender'] == 'M' else fake.first_name_female(), axis=1)
df_cust['lname'] = df_cust.apply(lambda x: fake.last_name_male() if x['gender'] == 'M' else fake.last_name_female(), axis=1)
# Using custom email generator function with numpy vectorize function, calling fname and lname columns in it
df_cust['email'] = np.vectorize(email_generator)(df_cust['fname'], df_cust['lname'])

df_cust
```
